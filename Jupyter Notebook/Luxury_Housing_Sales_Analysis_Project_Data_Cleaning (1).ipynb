```python
def detect_outliers_iqr(df, col):
  """Detects outliers in a column using the IQR method."""
  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)
  IQR = Q3 - Q1
  lower = Q1 - 1.5 * IQR
  upper = Q3 + 1.5 * IQR
  outliers = df[(df[col] < lower) | (df[col] > upper)]
  print(f"{col}: {len(outliers)} outliers detected")
  return lower, upper

num_cols = ['Unit_Size_Sqft',
            'Ticket_Price_Cr',
            'Connectivity_Score',
            'Amenity_Score',
            'Locality_Infra_Score',
            'Avg_Traffic_Time_Min']

for col in num_cols:
  lower, upper = detect_outliers_iqr(df, col)
```

***8. Feature Engineering***


```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Price per sqft

df['Price_Per_Sqft'] = (df['Ticket_Price_Cr'] * 10000000) / df['Unit_Size_Sqft']

# Split Purchase Quarter into year and quarter

df['Purchase_Year'] = df['Purchase_Quarter'].dt.year
df['Purchase_Qtr'] = df['Purchase_Quarter'].dt.quarter

# Overall Quality Score

df['Overall_Quality_Score'] = (df['Amenity_Score'] + df['Connectivity_Score'] + df['Locality_Infra_Score'] / 3)

# Traffic Efficiency Score

df['Traffic_Efficiency'] = df['Connectivity_Score'] /df['Avg_Traffic_Time_Min']

# Remove Buyer Comment Insights as the column was dropped
# df['Comment_Length'] = df['Buyer_Comments'].fillna('').apply(len)
# df['Has_Positive_Word'] = df['Buyer_Comments'].str.contains("excellent|good|great|loved|amazing|satisfied", case=False, na=False)

print("Derived Features Created Sucessfully")
```

```python
# Label Encoding for Binary Variables

le = LabelEncoder()
df['NRI_Buyer'] = le.fit_transform(df['NRI_Buyer'])
df['Transaction_Type'] = le.fit_transform(df['Transaction_Type'])

# One-Hot Encoding for Categorical Variables
columns_to_encode = ['Buyer_Type', 'Sales_Channel', 'Possession_Status']
# Check if columns exist before encoding
existing_columns_to_encode = [col for col in columns_to_encode if col in df.columns]
df = pd.get_dummies(df, columns=existing_columns_to_encode, drop_first=True)

print("Categorical Features Encoded")
```

```python
# Handle Skewed Columns

df['Ticket_Price_Log'] = np.log1p(df['Ticket_Price_Cr'])
df['Unit_Size_Log'] = np.log1p(df['Unit_Size_Sqft'])

print("Log Transformations Applied for Skewed Numeric Features")
```

```python
# Drop Irrevelant Columns

# df.drop(['Property_ID', 'Project_Name', 'Buyer_Comments'], axis=1, inplace=True)

print("Irrevelant Columns Removed")
```

```python
# Final Data Check

print("\n Final Dataset Summary:\n")
print(df.info())
print("\n Feature Engineering Completed Successfully")
```

```python
import pandas as pd

# Load both datasets
raw_df = pd.read_csv("Luxury_Housing_Bangalore.csv")
feature_df = pd.read_csv("Luxury_Housing_Feature_Engineered.csv")

df_sql = feature_df.copy()

restore_cols = ['Property_ID', 'Project_Name', 'Buyer_Comments',
                'Buyer_Type', 'Sales_Channel', 'Possession_Status']

for col in restore_cols:
    if col in raw_df.columns and col not in df_sql.columns:
        df_sql[col] = raw_df[col]

df_sql['Transaction_Type'] = df_sql['Transaction_Type'].replace({1: 'Primary', 0: 'Secondary'})
df_sql['NRI_Buyer'] = df_sql['NRI_Buyer'].replace({1: 'Yes', 0: 'No'})

drop_cols = [
    'Comment_Length', 'Has_Positive_Word',
    'Ticket_Price_Log', 'Unit_Size_Log',
    'Buyer_Type_HNI', 'Buyer_Type_NRI', 'Buyer_Type_Other', 'Buyer_Type_Startup Founder',
    'Sales_Channel_Direct', 'Sales_Channel_NRI Desk', 'Sales_Channel_Online',
    'Possession_Status_Ready to move', 'Possession_Status_Under construction'
]
df_sql.drop(columns=drop_cols, inplace=True, errors='ignore')

final_order = [
    'Property_ID', 'Project_Name', 'Developer_Name', 'Micro_Market', 'Configuration',
    'Unit_Size_Sqft', 'Ticket_Price_Cr', 'Price_Per_Sqft',
    'Transaction_Type', 'Buyer_Type', 'NRI_Buyer',
    'Purchase_Quarter', 'Purchase_Year', 'Purchase_Qtr',
    'Connectivity_Score', 'Amenity_Score', 'Locality_Infra_Score',
    'Overall_Quality_Score', 'Traffic_Efficiency', 'Avg_Traffic_Time_Min',
    'Sales_Channel', 'Possession_Status', 'Buyer_Comments'
]
df_sql = df_sql[[col for col in final_order if col in df_sql.columns]]

df_sql.to_csv("Luxury_Housing.csv", index=False, encoding='utf-8-sig')
print("Final dataset saved successfully as 'Luxury_Housing.csv'")

df_sql.head()
```
